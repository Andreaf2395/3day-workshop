# Polymorphism in Class

- Polymorphism is the ability to create a new object with varying attributes and common functionality of a class

- multiple instances or objects stemming from same class eg. Class car will have objects 
    - Hyundai-I20, 
    
    - Honda-Amaze, 

etc. 

# Class declaration (example):

```python
class Geometric_Shape():
    # member variables
    
    
    # constructor
    def __init__(self, name, num_edges, num_vertices, height, width, area):
        self.name = name
        self.num_edges = num_edges
        self.num_vertices = num_vertices
        self.height = 10
        self.width = 20
        self.area = 0


    
    # member method
    def Area_Shape(self):
        if(self.num_edges == 4):
            self.area = self.height * self.width
            return self.area
```

# Polymorphic usage:

```python
# creating a rectangular object
sample_rectangle = Geometric_Shape(name = "Rectangle",num_edges=4,num_vertices=4,height=15,width=30,area=0)

# creating a triangular object
sample_triangle = Geometric_Shape(name = "Triangle",num_edges=3,num_vertices=3,height=10,width=20,area=0)
```
